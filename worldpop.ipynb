```python
import pandas as pd

# Load the dataset
file_path = 'C:/Users/dipay/OneDrive/Desktop/world_population.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Bar Plot: Population of countries in 2020
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='2020 Population', y='Country/Territory', order=df.sort_values('2020 Population', ascending=False)['Country/Territory'])
plt.title('Population of Countries in 2020')
plt.xlabel('Population')
plt.ylabel('Country/Territory')
plt.show()

# Line Plot: Population trend over the years for a few selected countries
selected_countries = ['China', 'India', 'United States', 'Indonesia', 'Brazil']
population_columns = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']

# Reshape data for line plot
df_selected = df[df['Country/Territory'].isin(selected_countries)]
df_melted = df_selected.melt(id_vars=['Country/Territory'], value_vars=population_columns, var_name='Year', value_name='Population')
df_melted['Year'] = df_melted['Year'].str.extract('(\d+)').astype(int)

plt.figure(figsize=(12, 8))
sns.lineplot(data=df_melted, x='Year', y='Population', hue='Country/Territory')
plt.title('Population Trend of Selected Countries (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()

# Box Plot: Distribution of populations across continents in 2020
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Continent', y='2020 Population')
plt.title('Population Distribution Across Continents in 2020')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.show()
```
